```python
import requests
from bs4 import BeautifulSoup
```

```python
url="https://en.wikipedia.org/wiki/List_of_largest_banks"
page=requests.get(url)
```

```python
soup=BeautifulSoup(page.text,'html.parser')
```

```python
soup.prettify()
```

```python
for ele in soup.find_all('tr'):
  print(ele)
  print('\n'*2)
```

```python
lst=[]
for tag in soup.find_all('a'):
  lst.append(tag.text)
```

```python
for name in soup.find_all('td'):
  print(name)
  print('\n'*3)
```

```python
tr=soup.find_all('tr')
```

```python
data=[]
for each_tr in tr:
  td=each_tr.find_all('td')
  for each_td in td:
    data.append(each_td.text)

```

```python
data
```

```python
len(data)
```

```python
new_dt=data[359:]
new_dt
```

```python


bank=[]
rank=[]
mcap=[]
i=0
for i in range(0,150,3):
  bank.append(new_dt[i+1])
  rank.append(new_dt[i])
  mcap.append(new_dt[i+2])
```

```python
Bank=[]
Rank=[]
Mcap=[]
for ele in bank:
  Bank.append(ele.replace("\n","").strip())

for ele in rank:
  Rank.append(ele.replace("\n","").strip())


for ele in mcap:
  Mcap.append(ele.replace("\n","").strip())

Mcap[0]=400.37
```

```python
dict1={"RANK":Rank,"BANK NAME":Bank,"MARKET CAPITALIZATION in $BILLIONS":Mcap}
```

```python
import pandas as pd
df=pd.DataFrame.from_dict(dict1)
df.head()
```
